### 1. Percentual por sexo das pessoas 

```python
tabela_sexo_pessoas = tabela_pessoa['CO_SEXO_PESSOA'] #tabela apenas com a coluna contendo o sexo das pessoas
#display(tabela_sexo_pessoas)
contagem_pessoas_sexo = tabela_sexo_pessoas.value_counts() #conto quantas pessoas de cada sexo
#display(contagem_pessoas)

homens = 0
mulheres = 0

try:
    homens = contagem_pessoas_sexo[1]
except KeyError:
    print("Não há homens!")
try:
    mulheres = contagem_pessoas_sexo[2]
except KeyError:
    print("Não há mulheres!")
    
#print(homens, mulheres)

total_pessoas_sexo = homens + mulheres

display("Existem " + str(homens) + " pessoas do sexo masculino")
display("Existem " + str(mulheres) + " pessoas do sexo feminino")
display("De um total de " + str(total_pessoas_sexo) + " pessoas")

values_percentual_pessoas_sexo = [(float("{0:.2f}".format((homens/total_pessoas_sexo))))*100, (float("{0:.2f}".format((mulheres/total_pessoas_sexo))))*100]
labels = ['Homens', 'Mulheres']
plt.pie(values_percentual_pessoas_sexo, labels = values_percentual_pessoas_sexo)
plt.title('Sexo das Pessoas em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

#print(values_percentual_pessoas_sexo)
#print(labels)
df = pd.DataFrame([[homens, mulheres]], columns=labels)
#display(df)
barras_sexo_pessoas = df.plot.bar(rot=0)
barras_sexo_pessoas.set_xlabel("Sexo", fontsize=12)
barras_sexo_pessoas.set_ylabel("Quantidade de Pessoas", fontsize=12)
```

### 2. Faixas etárias por sexo das pessoas em famílias cadastradas de João Pessoa-PB

```python
def calculateAge(birthDate): 
    today = date.today() 
    age = today.year - birthDate.year - ((today.month, today.day) < (birthDate.month, birthDate.day)) 
  
    return age

pessoas_datas_nascimento = tabela_pessoa['DT_NASC_PESSOA'] #tabela com as datas de nascimento das pessoas
#display(pessoas_datas_nascimento.head(10))
total_pessoas_datas_nascimento = len(pessoas_datas_nascimento.index) #Guardo o total para fazer o percentual depois
pessoas_datas_nascimento_aux = pd.to_datetime(pessoas_datas_nascimento, format="%Y/%m/%d")
#display(pessoas_datas_nascimento_aux.head(10))

i=0;

for index in pessoas_datas_nascimento_aux:
    i = i + 1
    if index.year > (date.today()).year:
        ano_correto_aux = index.year - 2000
        ano_correto = 1900 + ano_correto_aux
        str_data_correta = "%d/%d/%d" % (index.day, index.month, ano_correto)
        #print(str_data_correta)
        data_correta = (pd.to_datetime(str_data_correta, format="%Y/%m/%d")).date()
        #print(data_correta)
        #print(pessoas_datas_nascimento_aux[i - 1])
        pessoas_datas_nascimento_aux[i - 1] = data_correta

#display(pessoas_datas_nascimento_aux)

pessoas_0_4_anos = 0
pessoas_5_6_anos = 0
pessoas_7_15_anos = 0
pessoas_16_17_anos = 0
pessoas_18_24_anos = 0
pessoas_25_34_anos = 0
pessoas_35_39_anos = 0
pessoas_40_44_anos = 0
pessoas_45_49_anos = 0
pessoas_50_54_anos = 0
pessoas_55_59_anos = 0
pessoas_60_64_anos = 0
pessoas_maior_65_anos = 0

#for index in pessoas_datas_nascimento_aux:
    #print(abs(calculateAge(index)))

for index in pessoas_datas_nascimento_aux:
    if (calculateAge(index)>= 0 and calculateAge(index)<= 4):
        pessoas_0_4_anos = pessoas_0_4_anos + 1
        #print("Foram encontradas:",pessoas_0_4_anos,"pessoas entre 0 e 4 anos")
        
    elif (calculateAge(index)>= 5 and calculateAge(index)<= 6):
        pessoas_5_6_anos = pessoas_5_6_anos + 1
        #print("Foram encontradas:",pessoas_5_6_anos,"pessoas entre 5 e 6 anos")
        
    elif (calculateAge(index)>= 7 and calculateAge(index)<= 15):
        pessoas_7_15_anos = pessoas_7_15_anos + 1
        #print("Foram encontradas:",pessoas_7_15_anos,"pessoas entre 7 e 15 anos")
        
    elif (calculateAge(index)>= 16 and calculateAge(index)<= 17):
        pessoas_16_17_anos = pessoas_16_17_anos + 1
        #print("Foram encontradas:",pessoas_16_17_anos,"pessoas entre 16 e 17 anos")
        
    elif (calculateAge(index)>= 18 and calculateAge(index)<= 24):
        pessoas_18_24_anos = pessoas_18_24_anos + 1
        #print("Foram encontradas:",pessoas_18_24_anos,"pessoas entre 18 e 24 anos")
        
    elif (calculateAge(index)>= 25 and calculateAge(index)<= 34):
        pessoas_25_34_anos = pessoas_25_34_anos + 1
        #print("Foram encontradas:",pessoas_25_34_anos,"pessoas entre 25 e 34 anos")
        
    elif (calculateAge(index)>= 35 and calculateAge(index)<= 39):
        pessoas_35_39_anos = pessoas_35_39_anos + 1
        #print("Foram encontradas:",pessoas_35_39_anos,"pessoas entre 35 e 39 anos")
        
    elif (calculateAge(index)>= 40 and calculateAge(index)<= 44):
        pessoas_40_44_anos = pessoas_40_44_anos + 1
        #print("Foram encontradas:",pessoas_40_44_anos,"pessoas entre 40 e 44 anos")
        
    elif (calculateAge(index)>= 45 and calculateAge(index)<= 49):
        pessoas_45_49_anos = pessoas_45_49_anos + 1
        #print("Foram encontradas:",pessoas_45_49_anos,"pessoas entre 45 e 49 anos")
        
    elif (calculateAge(index)>= 50 and calculateAge(index)<= 54):
        pessoas_50_54_anos = pessoas_50_54_anos + 1
        #print("Foram encontradas:",pessoas_50_54_anos,"pessoas entre 50 e 54 anos")
        
    elif (calculateAge(index)>= 55 and calculateAge(index)<= 59):
        pessoas_55_59_anos = pessoas_55_59_anos + 1
        #print("Foram encontradas:",pessoas_55_59_anos,"pessoas entre 55 e 59 anos")
        
    elif (calculateAge(index)>= 60 and calculateAge(index)<= 64):
        pessoas_60_64_anos = pessoas_60_64_anos + 1
        #print("Foram encontradas:",pessoas_60_64_anos,"pessoas entre 60 e 64 anos")
        
    elif (calculateAge(index)>= 65):
        pessoas_maior_65_anos = pessoas_maior_65_anos + 1
        #print("Foram encontradas:",pessoas_maior_65_anos,"pessoas maiores de 65 anos")
        
print("Foram encontradas:",pessoas_0_4_anos,"pessoas entre 0 e 4 anos")
print("Foram encontradas:",pessoas_5_6_anos,"pessoas entre 5 e 6 anos")
print("Foram encontradas:",pessoas_7_15_anos,"pessoas entre 7 e 15 anos")
print("Foram encontradas:",pessoas_16_17_anos,"pessoas entre 16 e 17 anos")
print("Foram encontradas:",pessoas_18_24_anos,"pessoas entre 18 e 24 anos")
print("Foram encontradas:",pessoas_25_34_anos,"pessoas entre 25 e 34 anos")
print("Foram encontradas:",pessoas_35_39_anos,"pessoas entre 35 e 39 anos")
print("Foram encontradas:",pessoas_40_44_anos,"pessoas entre 40 e 44 anos")
print("Foram encontradas:",pessoas_45_49_anos,"pessoas entre 45 e 49 anos")
print("Foram encontradas:",pessoas_50_54_anos,"pessoas entre 50 e 54 anos")
print("Foram encontradas:",pessoas_55_59_anos,"pessoas entre 55 e 59 anos")
print("Foram encontradas:",pessoas_60_64_anos,"pessoas entre 60 e 64 anos")
print("Foram encontradas:",pessoas_maior_65_anos,"pessoas maiores de 65 anos")


display("De um total de " + str(total_pessoas_datas_nascimento) + " pessoas")


valores_percentual_pessoas_datas_nascimento = [(float("{0:.2f}".format((pessoas_0_4_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_5_6_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_7_15_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_16_17_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_18_24_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_25_34_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_35_39_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_40_44_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_45_49_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_50_54_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_55_59_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_60_64_anos/total_pessoas_datas_nascimento))))*100, (float("{0:.2f}".format((pessoas_maior_65_anos/total_pessoas_datas_nascimento))))*100]
labels = ['Entre 0 a 4 anos', 'Entre 5 a 6 anos', 'Entre 7 a 15 anos', 'Entre 16 a 17 anos', 'Entre 18 a 24 anos', 'Entre 25 a 34 anos', 'Entre 35 a 39 anos', 'Entre 40 a 44 anos', 'Entre 45 a 49 anos', 'Entre 50 a 54 anos', 'Entre 55 a 59 anos', 'Entre 60 a 64 anos', 'Maiores que 65 anos']
plt.pie(valores_percentual_pessoas_datas_nascimento, labels = valores_percentual_pessoas_datas_nascimento)
plt.title('Pessoas por Faixa Etária em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(20,20) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[pessoas_0_4_anos, pessoas_5_6_anos, pessoas_7_15_anos, pessoas_16_17_anos, pessoas_18_24_anos, pessoas_25_34_anos, pessoas_35_39_anos, pessoas_40_44_anos, pessoas_45_49_anos, pessoas_50_54_anos, pessoas_55_59_anos, pessoas_60_64_anos, pessoas_maior_65_anos]], columns=labels)
#display(df)
barras_idade_pessoas = df.plot.bar(rot=0, figsize = (15,7))
barras_idade_pessoas.set_xlabel("Faixa etária", fontsize=12)
barras_idade_pessoas.set_ylabel("Quantidade de Pessoas", fontsize=12)

```

### 4. Qual o percentual de pessoas beneficiárias do Programa Bolsa Família?

```python
tabela_beneficiarios_BF = tabela_pessoa['MARC_PBF'] #tabela apenas com a coluna de beneficiários
#display(tabela_beneficiarios_BF)
contagem_beneficiarios = tabela_beneficiarios_BF.value_counts() #conto ocorrências de valores
#display(contagem_beneficiarios)

beneficiarios_BF = 0
nao_beneficiarios_BF = 0

try:
    beneficiarios_BF = contagem_beneficiarios[1]
except KeyError:
    print("Não há beneficiários do BF")
try:
    nao_beneficiarios_BF = contagem_beneficiarios[0]
except KeyError:
    print("Não há não beneficiários do BF")
    
#print (beneficiarios_BF)

total_coluna_BF = beneficiarios_BF + nao_beneficiarios_BF

display("Existem " + str(beneficiarios_BF) + " pessoas do beneficiárias do BF")
display("Existem " + str(nao_beneficiarios_BF) + " pessoas não beneficiárias do BF")
display("De um total de " + str(total_coluna_BF) + " pessoas")

values_percentual_BF = [(float("{0:.2f}".format((beneficiarios_BF/total_coluna_BF))))*100, (float("{0:.2f}".format((nao_beneficiarios_BF/total_coluna_BF))))*100]
labels = ['Pessoas Beneficiárias', 'Pessoas Não Beneficiárias']
plt.pie(values_percentual_BF, labels = values_percentual_BF)
plt.title('Pessoas Beneficiárias em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[beneficiarios_BF, nao_beneficiarios_BF]], columns=labels)
barras_pessoas_beneficiarias = df.plot.bar(rot=0, figsize = (15,7))
barras_pessoas_beneficiarias.set_xlabel("Pessoas Beneficiárias", fontsize=12)
barras_pessoas_beneficiarias.set_ylabel("Quantidade de Pessoas", fontsize=12)
```

### 5. Estado cadastral da pessoa no cadastro único?

```python
tabela_estado_cadastral_pessoa = tabela_pessoa['CO_EST_CADASTRAL_MEMB'] #tabela apenas com o estado cadastral
#display(tabela_estado_cadastral_pessoa)
contagem_estado_cadastral_pessoa = tabela_estado_cadastral_pessoa.value_counts() #conto ocorrências de valores
#display(contagem_estado_cadastral_pessoa)

cadastrado = 0
aguardando_atribuicao_nis = 0
aguardando_alteracao_caracterizacao = 0

try:
    cadastrado = contagem_estado_cadastral_pessoa[3]
except KeyError:
    print("Não há pessoas cadastradas")
try:
    aguardando_atribuicao_nis = contagem_estado_cadastral_pessoa[5]
except KeyError:
    print("Não há pessoas aguardando atribuição do NIS")
try:
    aguardando_alteracao_caracterizacao = contagem_estado_cadastral_pessoa[6]
except KeyError:
    print("Não há pessoas aguardando alteração de caracterização")
    

total_coluna_estado_cadastral_pessoa = cadastrado + aguardando_atribuicao_nis + aguardando_alteracao_caracterizacao
#display(total_coluna_estado_cadastral_pessoa)

display("Existem " + str(cadastrado) + " pessoas cadastradas")
display("Existem " + str(aguardando_atribuicao_nis) + " pessoas aguardando atribuição do NIS")
display("Existem " + str(aguardando_alteracao_caracterizacao) + " pessoas aguardando alteração de caracterização")
display("De um total de " + str(total_coluna_estado_cadastral_pessoa) + " pessoas")

values_percentual_estado_cadastral_pessoa = [(float("{0:.2f}".format((cadastrado/total_coluna_estado_cadastral_pessoa))))*100, (float("{0:.2f}".format((aguardando_atribuicao_nis/total_coluna_estado_cadastral_pessoa))))*100, (float("{0:.2f}".format((aguardando_alteracao_caracterizacao/total_coluna_estado_cadastral_pessoa))))*100]
labels = ['Pessoas Cadastradas', 'Pessoas Aguardando Atribuição NIS', 'Pessoas Aguardando Alteração de Caracterização']
plt.pie(values_percentual_estado_cadastral_pessoa, labels = values_percentual_estado_cadastral_pessoa)
plt.title('Estado Cadastral das Pessoas em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[cadastrado, aguardando_atribuicao_nis, aguardando_alteracao_caracterizacao]], columns=labels)
barras_estado_cadastral_pessoa = df.plot.bar(rot=0, figsize = (15,7))
barras_estado_cadastral_pessoa.set_xlabel("Pessoas por Estado Cadastral", fontsize=12)
barras_estado_cadastral_pessoa.set_ylabel("Quantidade de Pessoas", fontsize=12)
```

### 6. Distribuição de raça/cor das pessoas cadastradas:

```python
tabela_raca_cor_pessoa = tabela_pessoa['CO_EST_CADASTRAL_MEMB'] #tabela apenas com a raça/cor
#display(tabela_raca_cor_pessoa)
contagem_raca_cor_pessoa = tabela_raca_cor_pessoa.value_counts() #conto ocorrências de valores
#display(contagem_raca_cor_pessoa)

branca = 0
preta = 0
amarela = 0
parda = 0
indigena = 0

try:
    branca = contagem_raca_cor_pessoa[1]
except KeyError:
    print("Não há pessoas brancas")
try:
    preta = contagem_raca_cor_pessoa[2]
except KeyError:
    print("Não há pessoas pretas")
try:
    amarela = contagem_raca_cor_pessoa[3]
except KeyError:
    print("Não há pessoas amarelas")
try:
    parda = contagem_raca_cor_pessoa[4]
except KeyError:
    print("Não há pessoas pardas")
try:
    indigena = contagem_raca_cor_pessoa[5]
except KeyError:
    print("Não há pessoas indigenas")
    

total_coluna_raca_cor_pessoa = branca + preta + amarela + parda + indigena
#display(total_coluna_raca_cor_pessoa)

display("Existem " + str(branca) + " pessoas brancas")
display("Existem " + str(preta) + " pessoas pretas")
display("Existem " + str(amarela) + " pessoas amarelas")
display("Existem " + str(parda) + " pessoas pardas")
display("Existem " + str(indigena) + " pessoas indígenas")
display("De um total de " + str(total_coluna_raca_cor_pessoa) + " pessoas")

values_percentual_raca_cor_pessoa = [(float("{0:.2f}".format((branca/total_coluna_raca_cor_pessoa))))*100, (float("{0:.2f}".format((preta/total_coluna_raca_cor_pessoa))))*100, (float("{0:.2f}".format((amarela/total_coluna_raca_cor_pessoa))))*100, (float("{0:.2f}".format((parda/total_coluna_raca_cor_pessoa))))*100, (float("{0:.2f}".format((indigena/total_coluna_raca_cor_pessoa))))*100]
labels = ['Pessoas Brancas', 'Pessoas Pretas', 'Pessoas Amarelas', 'Pessoas Pardas', 'Pessoas Indígenas']
plt.pie(values_percentual_raca_cor_pessoa, labels = values_percentual_raca_cor_pessoa)
plt.title('Raça/Cor das Pessoas em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[branca, preta, amarela, parda, indigena]], columns=labels)
barras_raca_cor_pessoa = df.plot.bar(rot=0, figsize = (15,7))
barras_raca_cor_pessoa.set_xlabel("Pessoas por Raça/Cor", fontsize=12)
barras_raca_cor_pessoa.set_ylabel("Quantidade de Pessoas", fontsize=12)
```

### 15. Função principal - Identificar a relação de trabalho existente entre a pessoa e o empreendimento (negócio, firma, instituição, etc.) em que trabalha, o trabalho principal que a pessoa tinha na semana anterior à data da entrevista.

```python
tabela_pessoa_funcao = tabela_pessoa['CO_PRINCIPAL_TRAB_MEMB'] #tabela apenas com a coluna de trabalho
tabela_pessoa_funcao.fillna(0, inplace = True)
#display(tabela_pessoa_funcao)
contagem_pessoa_funcao = tabela_pessoa_funcao.value_counts() #conto ocorrências de valores
#display (contagem_pessoa_funcao)

trab_conta_prop = 0
trab_temp_rural = 0
trab_sem_carteira = 0
trab_com_carteira = 0
trab_dom_sem_carteira = 0
trab_dom_com_carteira = 0
trab_nao_remun = 0
trab_militar = 0
trab_empregador = 0
trab_estagiario = 0
trab_aprendiz = 0

try:
    trab_conta_prop = contagem_pessoa_funcao[1]
except KeyError:
    print("Não há pessoas que trabalham por conta própria")
try:
    trab_temp_rural = contagem_pessoa_funcao[2]
except KeyError:
    print("Não há pessoas que trabalham temporário rural")
try:
    trab_sem_carteira = contagem_pessoa_funcao[3]
except KeyError:
    print("Não há pessoas que trabalham sem carteira assinada")
try:
    trab_com_carteira = contagem_pessoa_funcao[4]
except KeyError:
    print("Não há pessoas que trabalham com carteira assinada")
try:
    trab_dom_sem_carteira = contagem_pessoa_funcao[5]
except KeyError:
    print("Não há pessoas que trabalham doméstico sem carteira assinada")
try:
    trab_dom_com_carteira = contagem_pessoa_funcao[6]
except KeyError:
    print("Não há pessoas que trabalham doméstico com carteira assinada")
try:
    trab_nao_remun = contagem_pessoa_funcao[7]
except KeyError:
    print("Não há pessoas que trabalham não remunerado")
try:
    trab_militar = contagem_pessoa_funcao[8]
except KeyError:
    print("Não há pessoas que trabalham militar ou servidor")
try:
    trab_empregador = contagem_pessoa_funcao[9]
except KeyError:
    print("Não há pessoas que trabalham como empregador")
try:
    trab_estagiario = contagem_pessoa_funcao[10]
except KeyError:
    print("Não há pessoas que trabalham como estagiário")
try:
    trab_aprendiz = contagem_pessoa_funcao[11]
except KeyError:
    print("Não há pessoas que trabalham como aprendiz")
    

total_coluna_pessoa_funcao = trab_conta_prop + trab_temp_rural + trab_sem_carteira + trab_com_carteira + trab_dom_sem_carteira + trab_dom_com_carteira + trab_nao_remun + trab_militar + trab_empregador + trab_estagiario + trab_aprendiz

display("Existem " + str(trab_conta_prop) + " pessoas que trabalham por conta própria")
display("Existem " + str(trab_temp_rural) + " pessoas em trabalho temporário rural")
display("Existem " + str(trab_sem_carteira) + " pessoas em trabalho sem carteira assinada")
display("Existem " + str(trab_com_carteira) + " pessoas em trabalho com carteira assinada")
display("Existem " + str(trab_dom_sem_carteira) + " pessoas em trabalho doméstico sem carteira assinada")
display("Existem " + str(trab_dom_com_carteira) + " pessoas em trabalho doméstico com carteira assinada")
display("Existem " + str(trab_nao_remun) + " pessoas em trabalho não remunerado")
display("Existem " + str(trab_militar) + " pessoas em trabalho militar ou servidor público")
display("Existem " + str(trab_empregador) + " pessoas em trabalho empregador")
display("Existem " + str(trab_estagiario) + " pessoas em trabalho de estagiário")
display("Existem " + str(trab_aprendiz) + " pessoas em trabalho de aprendiz")
display("De um total de " + str(total_coluna_pessoa_funcao) + " pessoas")

values_percentual_pessoa_funcao = [(float("{0:.2f}".format((trab_conta_prop/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_temp_rural/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_sem_carteira/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_com_carteira/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_dom_sem_carteira/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_dom_com_carteira/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_nao_remun/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_militar/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_empregador/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_estagiario/total_coluna_pessoa_funcao))))*100, (float("{0:.2f}".format((trab_aprendiz/total_coluna_pessoa_funcao))))*100]
labels = ['Pessoas que Trabalham por Conta Própria', 'Pessoas que Trabalham em Emprego Temporário Rural', 'Pessoas que Trabalham sem carteira assinada', 'Pessoas que Trabalham com carteira assinada', 'Pessoas que Trabalham doméstico e sem carteira assinada', 'Pessoas que Trabalham doméstico e com carteira assinada', 'Trabalhadores não remunerados', 'Militar ou servidor público', 'Empregador', 'Estagiário', 'Aprendiz']
plt.pie(values_percentual_pessoa_funcao, labels = values_percentual_pessoa_funcao)
plt.title('Principal atividade das pessoas (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[trab_conta_prop, trab_temp_rural, trab_sem_carteira, trab_com_carteira, trab_dom_sem_carteira, trab_dom_com_carteira, trab_nao_remun, trab_militar, trab_empregador, trab_estagiario, trab_aprendiz]], columns=labels)
barras_pessoas_funcao = df.plot.bar(rot=0, figsize = (15,7))
barras_pessoas_funcao.set_xlabel("Principal atividade das pessoas", fontsize=12)
barras_pessoas_funcao.set_ylabel("Quantidade de Pessoas", fontsize=12)
```
