### 1. Qual o percentual da situação econômica de Famílias por faixa de Renda – João Pessoa-PB?

```python
valor_salario_min = 998.00

#Tabela apenas com as famílias em extrema pobreza
extrema_pobreza_bool = tabela_familia['VL_RENDA_MEDIA_FAM']<=89 #booleano categoriza famílias com essa renda
extrema_pobreza = tabela_familia[extrema_pobreza_bool] #Filtra tabela original pelo booleano
print('Famílias em Extrema Pobreza')
display(extrema_pobreza.head(10)) #Exibe amostra da tabela

#Tabela apenas com as famílias em pobreza
pobreza_bool = tabela_familia['VL_RENDA_MEDIA_FAM']<=178 #booleano categoriza famílias com essa renda
pobreza_aux = tabela_familia[pobreza_bool] #Filtra tabela original pelo booleano
pobreza_bool_2 = pobreza_aux['VL_RENDA_MEDIA_FAM']>89 #Segunda fase da categorização
pobreza = pobreza_aux[pobreza_bool_2] #Segunda fase do filtro para gerar a tabela final
print('\n\n\n\nFamílias em Pobreza')
display(pobreza.head(10))

#Tabela apenas com as famílias em baixa renda 1
baixa_renda_1_bool = tabela_familia['VL_RENDA_MEDIA_FAM']>178 #booleano categoriza famílias com essa renda
baixa_renda_1_aux = tabela_familia[baixa_renda_1_bool] #Filtra tabela original pelo booleano
baixa_renda_1_bool_2 = baixa_renda_1_aux['VL_RENDA_MEDIA_FAM']<=(valor_salario_min / 2) #Segunda fase da categorização
baixa_renda_1 = baixa_renda_1_aux[baixa_renda_1_bool_2] #Segunda fase do filtro para gerar a tabela final
print('\n\n\n\nFamílias em Baixa Renda 1')
display(baixa_renda_1.head(10))

#Tabela apenas com as famílias em baixa renda 2
baixa_renda_2_bool = tabela_familia['VL_RENDA_MEDIA_FAM']>(valor_salario_min / 2)
baixa_renda_2 = tabela_familia[baixa_renda_2_bool]
print('\n\n\n\nFamílias em Baixa Renda 2')
display(baixa_renda_2.head(10))

#Variáveis guardam quantas famílias existem em cada uma das tabelas auxiliares
familias_extrema_pobreza = len(extrema_pobreza.index)
familias_pobreza = len(pobreza.index)
familias_baixa_renda_1 = len(baixa_renda_1.index)
familias_baixa_renda_2 = len(baixa_renda_2.index)

total_familias_renda = familias_extrema_pobreza + familias_pobreza + familias_baixa_renda_1 + familias_baixa_renda_2

display("Existem " + str(familias_extrema_pobreza) + " famílias em situação de extrema pobreza")
display("Existem " + str(familias_pobreza) + " famílias em situação de pobreza")
display("Existem " + str(familias_baixa_renda_1) + " famílias em situação de baixa renda 1")
display("Existem " + str(familias_baixa_renda_2) + " famílias em situação de baixa renda 2")
display("De um total de " + str(total_familias_renda) + " famílias")

#Para a exibição do gráfico em números
values = [familias_extrema_pobreza, familias_pobreza, familias_baixa_renda_1, familias_baixa_renda_2]
labels = ['Extrema Pobreza', 'Pobreza', 'Baixa Renda 1', 'Baixa Renda 2']
plt.pie(values, labels = values)
plt.title('Famílias por Situação de Renda em Números')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

#Para a exibição do gráfico em percentual
total = familias_extrema_pobreza + familias_pobreza + familias_baixa_renda_1 + familias_baixa_renda_2
values_percentual_total = [(float("{0:.2f}".format((familias_extrema_pobreza/total))))*100, (float("{0:.2f}".format((familias_pobreza/total))))*100, (float("{0:.2f}".format((familias_baixa_renda_1/total))))*100, (float("{0:.2f}".format((familias_baixa_renda_2/total))))*100]
labels = ['Extrema Pobreza', 'Pobreza', 'Baixa Renda 1', 'Baixa Renda 2']
plt.pie(values_percentual_total, labels = values_percentual_total)
plt.title('Famílias por Situação de Renda em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[familias_extrema_pobreza, familias_pobreza, familias_baixa_renda_1, familias_baixa_renda_2]], columns=labels)
barras_faixa_renda = df.plot.bar(rot=0, figsize = (15,7))
barras_faixa_renda.set_xlabel("Famílias por Faixa de Renda", fontsize=12)
barras_faixa_renda.set_ylabel("Quantidade de Famílias", fontsize=12)

#df = pd.DataFrame({'Famílias por Situação de Renda': [familias_extrema_pobreza, familias_pobreza, familias_baixa_renda_1, familias_baixa_renda_2]}, index=['Extrema Pobreza', 'Pobreza', 'Baixa Renda 1', 'Baixa Renda 2'])
#plot = df.plot.pie(y='Famílias por Situação de Renda', labels=['Extrema Pobreza', 'Pobreza', 'Baixa Renda 1', 'Baixa Renda 2'] , figsize=(10,10))
```

### 2. Qual o percentual de Famílias por renda per capita – João Pessoa? 

```python
#Tabela apenas com as famílias que ganham menos de 1 salário mínimo
familias_menos_salario_min_bool = tabela_familia['VL_RENDA_MEDIA_FAM']<valor_salario_min #booleano categoriza famílias com essa renda
familias_menos_salario_min = tabela_familia[familias_menos_salario_min_bool] #Filtra tabela original pelo booleano
print('Famílias que Ganham Menos de Um Salário Mínimo')
display(familias_menos_salario_min.head(10)) #Exibe amostra da tabela


#Tabela apenas com as famílias que ganham exatamente 1 salário mínimo
familias_salario_min_bool = tabela_familia['VL_RENDA_MEDIA_FAM']==valor_salario_min #booleano categoriza famílias com essa renda
familias_salario_min = tabela_familia[familias_salario_min_bool] #Filtra tabela original pelo booleano
print('\n\n\n\nFamílias que Ganham Exatamente Um Salário Mínimo')
display(familias_salario_min.head(10)) #Exibe amostra da tabela


#Tabela com as famílias que ganham mais que 1 salário mínimo
familias_maior_salario_min_bool = tabela_familia['VL_RENDA_MEDIA_FAM']>valor_salario_min #booleano categoriza famílias com essa renda
familias_maior_salario_min = tabela_familia[familias_maior_salario_min_bool] #Filtra tabela original pelo booleano
print('\n\n\n\nFamílias que Ganham Mais de Um Salário Mínimo')
display(familias_maior_salario_min.head(10)) #Exibe amostra da tabela

#Variáveis guardam quantas famílias existem em cada uma das tabelas auxiliares
familias_menos_salario_min_count = len(familias_menos_salario_min.index)
familias_salario_min_count = len(familias_salario_min.index)
familias_mais_salario_min_count = len(familias_maior_salario_min.index)

#Para a exibição do gráfico em percentual
total2 = familias_menos_salario_min_count + familias_mais_salario_min_count + familias_salario_min_count

display("Existem " + str(familias_menos_salario_min_count) + " famílias que ganham menos que um salário mínimo")
display("Existem " + str(familias_mais_salario_min_count) + " famílias que ganham mais que um salário mínimo")
display("Existem " + str(familias_salario_min_count) + " famílias que ganham um salário mínimo")
display("De um total de " + str(total2) + " famílias")


values_percentual_total2 = [(float("{0:.2f}".format((familias_menos_salario_min_count/total2))))*100, (float("{0:.2f}".format((familias_mais_salario_min_count/total2))))*100, (float("{0:.2f}".format((familias_salario_min_count/total2))))*100]
labels = ['Famílias que recebem menos que um salário mínimo', 'Famílias que recebem mais de um salário mínimo', 'Famílias que recebem exatamente um salário mínimo']
plt.pie(values_percentual_total2, labels = values_percentual_total2)
plt.title('Famílias por Situação de Renda em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[familias_menos_salario_min_count, familias_mais_salario_min_count, familias_salario_min_count]], columns=labels)
barras_renda_percapita = df.plot.bar(rot=0, figsize = (15,7))
barras_renda_percapita.set_xlabel("Famílias por Renda per Capita", fontsize=12)
barras_renda_percapita.set_ylabel("Quantidade de Famílias", fontsize=12)
```

### 3. Qual o percentual de Famílias beneficiárias do Programa Bolsa Família?

```python
tabela_beneficiarios_BF = tabela_familia['MARC_PBF'] #tabela apenas com a coluna de beneficiários
#display(tabela_beneficiarios_BF)
contagem_beneficiarios = tabela_beneficiarios_BF.value_counts() #conto ocorrências de valores
#display(contagem_beneficiarios)

beneficiarios_BF = 0
nao_beneficiarios_BF = 0

try:
    beneficiarios_BF = contagem_beneficiarios[1]
except KeyError:
    print("Não há beneficiários do BF")
try:
    nao_beneficiarios_BF = contagem_beneficiarios[0]
except KeyError:
    print("Não há não beneficiários do BF")
    
#print (beneficiarios_BF)

total_coluna_BF = beneficiarios_BF + nao_beneficiarios_BF

display("Existem " + str(beneficiarios_BF) + " famílias beneficiárias do BF")
display("Existem " + str(nao_beneficiarios_BF) + " famílias não beneficiárias do BF")
display("De um total de " + str(total_coluna_BF) + " famílias")

values_percentual_BF = [(float("{0:.2f}".format((beneficiarios_BF/total_coluna_BF))))*100, (float("{0:.2f}".format((nao_beneficiarios_BF/total_coluna_BF))))*100]
labels = ['Famílias Beneficiárias', 'Famílias Não Beneficiárias']
plt.pie(values_percentual_BF, labels = values_percentual_BF)
plt.title('Famílias Beneficiárias em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[beneficiarios_BF, nao_beneficiarios_BF]], columns=labels)
barras_familias_beneficiarias = df.plot.bar(rot=0, figsize = (15,7))
barras_familias_beneficiarias.set_xlabel("Famílias Beneficiárias", fontsize=12)
barras_familias_beneficiarias.set_ylabel("Quantidade de Famílias", fontsize=12)
```

### 5. Percentual de família por Unidade Territorial Local no município de João Pessoa.

```python
#Fiz usando apenas códigos (CO_UNIDADE_TERRITORIAL_FAM), pois garante que estão iguais no banco, os nomes de uma localidade com o mesmo código podem divergir!

tabela_codigo_localidade = tabela_familia['CO_UNIDADE_TERRITORIAL_FAM'] #Apenas a coluna com os códigos
tabela_codigo_localidade.fillna(0, inplace = True) #Quando não preenchido, o código é 0, já que 0 não existe no dataset
#display(tabela_codigo_localidade)
total_familias_codigo_localidade = len(tabela_familia.index) #Faço a contagem do número total de famílias para a porcentagem final



array_codigo_localidade = tabela_codigo_localidade.astype(int).to_numpy() #Transformo em array para usar na legenda do gráfico
array_codigo_localidade.astype(int) #Converto para inteiro, para não perder os códigos
array_codigo_localidade = np.unique(array_codigo_localidade) #Removo duplicatas

#display(array_codigo_localidade)
contagem_familias_codigo_localidade = tabela_codigo_localidade.value_counts() #Conto quantas famílias tem em cada código de localidade
#display(contagem_familias_codigo_localidade)

array_contagem_familias_codigo_localidade = contagem_familias_codigo_localidade.to_numpy() #Converto para um array de inteiros

array_porcentagem_familias_codigo_localidade = (array_contagem_familias_codigo_localidade/total_familias_codigo_localidade*100) #Faço a porcentagem de famílias em cada localidade

print('\n\n Um total de ' + str(total_familias_codigo_localidade) + ' famílias estão distribuídas nos seguintes códigos de localidade:\n\n')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(contagem_familias_codigo_localidade)

#Desenho o gráfico
labels = array_codigo_localidade
plt.pie(array_porcentagem_familias_codigo_localidade, labels = array_porcentagem_familias_codigo_localidade)
plt.title('Famílias Por Código de Unidade Territorial Local em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(25,25) #tamanho do gráfico
plt.show()

df = pd.DataFrame([array_contagem_familias_codigo_localidade], columns=labels)
barras_familia_codigo_localidade = df.plot.bar(rot=0, figsize = (15,7))
barras_familia_codigo_localidade.set_xlabel("Famílias por Código de Unidade Territorial", fontsize=12)
barras_familia_codigo_localidade.set_ylabel("Quantidade de Famílias", fontsize=12)
```

### 6.  Características do local onde está situado o domicílio (autodeclarado, não oficial) em percentual:

```python
tabela_caracteristicas_local = tabela_familia['CO_LOCAL_DOMIC_FAM'] #tabela apenas com a coluna das características do local
#display(tabela_caracteristicas_local)

contagem_domicilios_caracteristica = tabela_caracteristicas_local.value_counts() #conto ocorrências de valores
#display(contagem_domicilios_caracteristica)

urbanos = 0
rurais = 0

try:
    urbanos = contagem_domicilios_caracteristica[1]
except KeyError:
    print("Não há domicílios urbanos")
try:
    rurais = contagem_domicilios_caracteristica[2]
except KeyError:
    print("Não há domicílios rurais")
    

#print(urbanos)
#print(rurais)

total_domicilios_caracteristica = urbanos + rurais

display("Existem " + str(urbanos) + " famílias em domicílios urbanos")
display("Existem " + str(rurais) + " famílias em domicílios rurais")
display("De um total de " + str(total_domicilios_caracteristica) + " famílias")

values_percentual_domicilios_caracteristica = [(float("{0:.2f}".format((urbanos/total_domicilios_caracteristica))))*100, (float("{0:.2f}".format((rurais/total_domicilios_caracteristica))))*100]
labels = ['Domicílios em local urbano', 'Domicílios em local rural']
plt.pie(values_percentual_domicilios_caracteristica, labels = values_percentual_domicilios_caracteristica)
plt.title('Característica da Localidade dos Domicílios em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[urbanos, rurais]], columns=labels)
barras_caracteristicas_local_familia = df.plot.bar(rot=0, figsize = (15,7))
barras_caracteristicas_local_familia.set_xlabel("Famílias por Característica da Localidade", fontsize=12)
barras_caracteristicas_local_familia.set_ylabel("Quantidade de Famílias", fontsize=12)
```

### 15. Classificação das famílas por classes econômicas sociais

```python
#Tabela apenas com as famílias na classe E
classe_e_bool = tabela_familia['VL_RENDA_MEDIA_FAM']<=1254 #booleano categoriza famílias com essa renda
classe_e = tabela_familia[classe_e_bool] #Filtra tabela original pelo booleano
#display(classe_e.head(10)) #Exibe amostra da tabela

#Tabela apenas com as famílias na classe D
classe_d_bool_1 = tabela_familia['VL_RENDA_MEDIA_FAM']>1254 #booleano categoriza famílias com essa renda
classe_d_aux = tabela_familia[classe_d_bool_1] #Filtra tabela original pelo booleano
classe_d_bool_2 = classe_d_aux['VL_RENDA_MEDIA_FAM']<=2004 #Segunda fase da categorização
classe_d = classe_d_aux[classe_d_bool_2] #Segunda fase do filtro para gerar a tabela final
#display(classe_d.head(10))

#Tabela apenas com as famílias na classe C
classe_c_bool_1 = tabela_familia['VL_RENDA_MEDIA_FAM']>2004 #booleano categoriza famílias com essa renda
classe_c_aux = tabela_familia[classe_c_bool_1] #Filtra tabela original pelo booleano
classe_c_bool_2 = classe_c_aux['VL_RENDA_MEDIA_FAM']<=8640 #Segunda fase da categorização
classe_c = classe_c_aux[classe_c_bool_2] #Segunda fase do filtro para gerar a tabela final
#display(classe_c.head(10))

#Tabela apenas com as famílias na classe B
classe_b_bool_1 = tabela_familia['VL_RENDA_MEDIA_FAM']>8640 #booleano categoriza famílias com essa renda
classe_b_aux = tabela_familia[classe_b_bool_1] #Filtra tabela original pelo booleano
classe_b_bool_2 = classe_b_aux['VL_RENDA_MEDIA_FAM']<=11261 #Segunda fase da categorização
classe_b = classe_b_aux[classe_b_bool_2] #Segunda fase do filtro para gerar a tabela final
#display(classe_b.head(10))

#Tabela apenas com as famílias na classe A
classe_a_bool = tabela_familia['VL_RENDA_MEDIA_FAM']>11261 #booleano categoriza famílias com essa renda
classe_a = tabela_familia[classe_a_bool] #Filtra tabela original pelo booleano
#display(classe_a.head(10)) #Exibe amostra da tabela

#Variáveis guardam quantas famílias existem em cada uma das tabelas auxiliares
familias_classe_e = len(classe_e.index)
familias_classe_d = len(classe_d.index)
familias_classe_c = len(classe_c.index)
familias_classe_b = len(classe_b.index)
familias_classe_a = len(classe_a.index)

total_familias_faixa_renda = familias_classe_e + familias_classe_d + familias_classe_c + familias_classe_b + familias_classe_a

display("Existem " + str(familias_classe_e) + " famílias classe E")
display("Existem " + str(familias_classe_d) + " famílias classe D")
display("Existem " + str(familias_classe_c) + " famílias classe C")
display("Existem " + str(familias_classe_b) + " famílias classe B")
display("Existem " + str(familias_classe_a) + " famílias classe A")
display("De um total de " + str(total_familias_faixa_renda) + " famílias")

#Para a exibição do gráfico em números
values = [familias_classe_e, familias_classe_d, familias_classe_c, familias_classe_b, familias_classe_a]
labels = ['Classe E', 'Classe D', 'Classe C', 'Classe B', 'Classe A']
plt.pie(values, labels = values)
plt.title('Famílias por Classe de Renda em Números')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

#Para a exibição do gráfico em percentual
total = total_familias_faixa_renda
values_percentual_total = [(float("{0:.2f}".format((familias_classe_e/total))))*100, (float("{0:.2f}".format((familias_classe_d/total))))*100, (float("{0:.2f}".format((familias_classe_c/total))))*100, (float("{0:.2f}".format((familias_classe_b/total))))*100, (float("{0:.2f}".format((familias_classe_a/total))))*100]
labels = ['Classe E', 'Classe D', 'Classe C', 'Classe B', 'Classe A']
plt.pie(values_percentual_total, labels = values_percentual_total)
plt.title('Famílias por Classe de Renda em Percentual (%)')
plt.legend(labels)
fig = plt.gcf()
fig.set_size_inches(10,10) #tamanho do gráfico
plt.show()

df = pd.DataFrame([[familias_classe_e, familias_classe_d, familias_classe_c, familias_classe_b, familias_classe_a]], columns=labels)
barras_classe_renda = df.plot.bar(rot=0, figsize = (15,7))
barras_classe_renda.set_xlabel("Famílias por Classe de Renda", fontsize=12)
barras_classe_renda.set_ylabel("Quantidade de Famílias", fontsize=12)
```
